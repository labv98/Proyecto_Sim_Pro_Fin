### 1.3 Definición del problema.


El aire es el recurso vital más importante para los seres humanos y su contaminación  que cada vez aumenta más es lo que más puede dañar nuestra salud y vida.
Autos, camiones , fábricas, quema de bosques y otros factores influyen en la calidad de aire que respiran los tapatíos.

El Índice Métropolinato de la Calidad del Aire (IMECA) permite conocer las condiciones del aire que se respira; para 2018 Guadalajara llegó a superar a la Ciudad de México con más de 150 puntos en dicho índice.
 Éste se calcula para cinco de los contaminantes criterio: 
 
* Dióxido de azufre

* Monóxido de carbono

* Dióxido de nitrógeno

* Ozono 

-partículas suspendidas

Se representa con una escala que va de 0 a 500, donde el valor de 100 se asigna al valor indicado por la Norma Oficial Mexicana para cada contaminante. Un valor menor a 100 se considera satisfactorio y con un bajo riesgo para la salud. Cualquier nivel superior a 100 implica algún riesgo para la salud

El deterioro de la calidad del aire por la presencia contaminantes tiene un efecto negativo en la salud humana aumentando la cantidad de enfermedades respiratorias y cardiovasculares además de afectar al medio ambiente. 

Debido a esto, se decidió realizar una simulación en la que se pueda estimar la calidad de aire específicamente analizando el monóxido de carbono (CO) y cómo éste se relaciona con la temperatura.

Desde finales del siglo XIX Svante Arrhenius encontró que los cambios en la temperatura promedio de la Tierra tienen una relación directa con los gases que hay en la atmósfera, en particular con los gases de efecto invernadero. Hoy se sabe con certeza que el aumento del CO2 genera un incremento en la radiación infrarroja cerca de la superficie de la Tierra y, por lo tanto, en la temperatura global del planeta

Debido a esto, el supuesto con el que se trabajará el proyecto es el hecho que el CO y la temperatura están correlacionados.

```python
from sklearn.model_selection import GridSearchCV
grid_co_aguilas = GridSearchCV(KernelDensity(),
                    {'bandwidth': np.linspace(0, 4.4, 30)},
                    cv=20) # 20-fold cross-validation

grid_co_centro = GridSearchCV(KernelDensity(),
                    {'bandwidth': np.linspace(0, 4.4, 30)},
                    cv=20) # 20-fold cross-validation

grid_co_aguilas.fit(co_aguilas[:, None])
grid_co_centro.fit(co_centro[:, None])

print(grid_co_aguilas.best_params_)
print(grid_co_centro.best_params_)
```

```python
def kde_sklearn(x, x_grid, bandwidth):
    """Kernel Density Estimation with Scikit-learn"""
    kde_skl = KernelDensity(bandwidth=bandwidth)
    kde_skl.fit(x[:, np.newaxis])
    # score_samples() returns the log-likelihood of the samples
    log_pdf = kde_skl.score_samples(x_grid[:, np.newaxis])
    return np.exp(log_pdf)
```

```python
x_grid = np.linspace(0, 4.4, 1000)

plt.plot(x_grid, kde_sklearn(co_aguilas, x_grid, bandwidth= 0.1517), label='Grid_Aguilas')
plt.plot(x_grid, kde_sklearn(co_centro, x_grid, bandwidth= 0.1517), label='Grid_Centro')
plt.legend(loc='best')
```

#### Pruebas de independencia

```python
a=np.asarray(co_aguilas)
b=np.asarray(temp_aguilas)
```

```python
est=[]
def chi(co,temp):#como arreglos
    for i in range(len(co)):
        x=co.sum()
        y=temp.sum()
        est.append(((co[i]+temp[i])*x)/x+y)
        est.append(((co[i]+temp[i])*y)/x+y)
    return est
```

```python
datos_est=chi(a,b)
obs=co_aguilas.append(temp_aguilas)
```

```python
#### Método chi cuadrado
x_aguilas = st.chisquare(obs,est,ddof=1)
print('Valor de chi cuadrado para las aguilas = ',list(x_aguilas)[0],',p-value de la prueba=',list(x_aguilas)[1])
Ji_aguilas = st.chi2.ppf(q = 0.9,df=1)
print('Estadístico de Ji para las aguilas = ',Ji_aguilas)

##############
x_centro = st.chisquare(co_centro,temp_centro,ddof=1)
print('Valor de chi cuadrado para centro = ',list(x_centro)[0],',p-value de la prueba=',list(x_centro)[1])
Ji_centro = st.chi2.ppf(q = 0.9,df=1)
print('Estadístico de Ji para centro = ',Ji_centro)
```

Rechazamos $H_a=$ No son independientes, Aceptamos $H_0=$ Son independientes

* Sabemos que el clima es un factor independiente a el CO pues siempre se va a tener una medida de temperatura la cual no necesita del Co para ser establecida sin embargo, podemos hacer otra prueba como la correlación que nos va a indicar si éstas dos variables pueden esar relacionadas entre sí.

```python
import seaborn as sns 
corr_aguilas = aguilas.corr() 
corr_centro = centro.corr()
corr_aguilas
```

```python
corr_centro
```

Al construir la matriz de correlación podemos observar que se tienen una correlación negativa; el ser negativa no significa que no estén relacionadas, si no que están relacionadas inversamente en un 36.17% y un 9.35% respectivamente para cada zona simulada. Esto de manera inversamente proporcional lo cual nos dice que a medida que una aumenta la otra baja. Es decir, a medida que disminuye la temperatura aumenta la cantidad de CO presente.

Cabe destacar que los datos nos dicen que son inversamente proporcionales cuando previamente habíamos investigado sobre el tema y encontramos que la relación entre CO y temperatura es muy alta, pero debemos recodar que ésta partícula no es la única que afecta el clima de la ciudad y que también la zona de muestra depende mucho de los comportamientos presentes, pues no es lo mismo las partículas que viajan por el aire en una zona industrial a las que viajan a una zona urbana.
Por esta razón, no descartamos la relación directa que habíamos encontrado, sino que esta partícula en conjunto con otras las cuales se encuentran presentes en el ambiente, provocarán mayor temperatura.

#### SIMULACIÓN MONTECARLO con Temperaturas

```python
aguilas['return'] = aguilas['Temp Ext'].pct_change().fillna(0)
mc_aguilas = aguilas['return'].montecarlo(sims=100, bust=-0.1, goal=1)
mc_aguilas.plot(title="Simulación Montecarlo Temp_AGUILAS")
```

```python
centro['return'] = centro['Temp Ext'].pct_change().fillna(0)
mc_centro = centro['return'].montecarlo(sims=100, bust=-0.1, goal=1)
mc_centro.plot(title="Simulación Montecarlo Temp_CENTRO")
```
