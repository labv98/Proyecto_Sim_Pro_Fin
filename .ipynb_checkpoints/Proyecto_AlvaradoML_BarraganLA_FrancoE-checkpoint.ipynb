### 1.3 Definición del problema.


El aire es el recurso vital más importante para los seres humanos y su contaminación  que cada vez aumenta más es lo que más puede dañar nuestra salud y vida.
Autos, camiones , fábricas, quema de bosques y otros factores influyen en la calidad de aire que respiran los tapatíos.

El Índice Métropolinato de la Calidad del Aire (IMECA) permite conocer las condiciones del aire que se respira; para 2018 Guadalajara llegó a superar a la Ciudad de México con más de 150 puntos en dicho índice.
 Éste se calcula para cinco de los contaminantes criterio: 
 
* Dióxido de azufre

* Monóxido de carbono

* Dióxido de nitrógeno

* Ozono 

-partículas suspendidas

Se representa con una escala que va de 0 a 500, donde el valor de 100 se asigna al valor indicado por la Norma Oficial Mexicana para cada contaminante. Un valor menor a 100 se considera satisfactorio y con un bajo riesgo para la salud. Cualquier nivel superior a 100 implica algún riesgo para la salud

El deterioro de la calidad del aire por la presencia contaminantes tiene un efecto negativo en la salud humana aumentando la cantidad de enfermedades respiratorias y cardiovasculares además de afectar al medio ambiente. 

Debido a esto, se decidió realizar una simulación en la que se pueda estimar la calidad de aire específicamente analizando el monóxido de carbono (CO) y cómo éste se relaciona con la temperatura.

Desde finales del siglo XIX Svante Arrhenius encontró que los cambios en la temperatura promedio de la Tierra tienen una relación directa con los gases que hay en la atmósfera, en particular con los gases de efecto invernadero. Hoy se sabe con certeza que el aumento del CO2 genera un incremento en la radiación infrarroja cerca de la superficie de la Tierra y, por lo tanto, en la temperatura global del planeta

Debido a esto, el supuesto con el que se trabajará el proyecto es el hecho que el CO y la temperatura están correlacionados.

```python
# Bases de datos
aguilas = pd.read_csv('aguilas.csv').dropna()
centro = pd.read_csv('centro.csv').dropna()
tlaquepaque = pd.read_csv('tlaquepaque.csv').dropna()
vallarta = pd.read_csv('vallarta.csv').dropna()
```

>### Histogramas y Kernels CO

```python
## Función para graficar histogramas
def subplot_dist(lis: 'Insertar la lista con los datos',titu: 'Insertar lista de títulos'):   
    fig, ax = plt.subplots(1, len(lis), sharey=True, figsize=(20, 6))
    fig.subplots_adjust(wspace=.1)
    [ax[i].hist(lis[i]) for i in range(len(lis))]
    [ax[i].set_title(titu[i]) for i in range(len(titu))]
```

```python
lista_co = [aguilas['CO'].values, centro['CO'].values, vallarta['CO'].values, tlaquepaque['CO'].values]
titulos_co = ['CO aguilas','CO centro', 'CO vallarta', 'CO tlaquepaque']
subplot_dist(lista_co,titulos_co)
```

En base a los histogramas realizados, podemos observar que la cantidad de CO presente en cada zona es distinta, pero parecen mostrar una misma distribución. 

Para encontrar la distribución que siguen nuestros datos, debemos encontrar el kernel de cada una de nuestras variables de estudio. Como cada zona presenta distintos máximos, es necesario realizar distintos rangos para tener una mejor presición.

```python
#GRIDS CO

#CO_AGUILAS
grid_ag = GridSearchCV(KernelDensity(),
                    {'bandwidth': np.linspace(0, 4.4, 50)},
                    cv=20)
grid_ag.fit(lista_co[0][:, None])
x_grid_ag = np.linspace(0, 4.4, 1000)
kde_ag = grid_ag.best_estimator_
pdf_ag = np.exp(kde_ag.score_samples(x_grid_ag[:, None])) #Kernel

#CENTRO 
grid_ce = GridSearchCV(KernelDensity(),
                    {'bandwidth': np.linspace(0, 4.4, 50)},
                    cv=20)
grid_ce.fit(lista_co[1][:, None])
x_grid_ce = np.linspace(0, 4.4, 1000)
kde_ce = grid_ce.best_estimator_
pdf_ce = np.exp(kde_ce.score_samples(x_grid_ce[:, None])) #Kernel

#VALLARTA
grid_va = GridSearchCV(KernelDensity(),
                    {'bandwidth': np.linspace(0, 4.4, 50)},
                    cv=20)
grid_va.fit(lista_co[2][:, None])
x_grid_va = np.linspace(0, 4.4, 1000)
kde_va = grid_va.best_estimator_
pdf_va = np.exp(kde_va.score_samples(x_grid_va[:, None])) #Kernel

#CO_TLAQUEPAQUE

grid_ta = GridSearchCV(KernelDensity(),
                    {'bandwidth': np.linspace(0, 11, 50)},
                    cv=20)
grid_ta.fit(lista_co[3][:, None])
x_grid_ta = np.linspace(0, 11, 1000)
kde_ta = grid_ta.best_estimator_
pdf_ta = np.exp(kde_ta.score_samples(x_grid_ta[:, None])) #Kernel
```

```python
## Función para graficar el kernel
def trans_a_kernel(x_grid,pdf,kde,datos,titulo):
    
    plt.figure()
    plt.plot(x_grid, pdf, linewidth=3, alpha=0.5, label='bw=%.2f' % kde.bandwidth)
    plt.hist(datos, 30, fc='gray', histtype='stepfilled', alpha=0.3, density=True) 
    plt.legend(loc='upper left')
    plt.title(titulo)
    plt.show()
```

```python
trans_a_kernel(x_grid_ag,pdf_ag,kde_ag,lista_co[0],titulos_co[0])
```

```python
trans_a_kernel(x_grid_ce,pdf_ce,kde_ce,lista_co[1],titulos_co[1])
```

> ### Aceptación y rechazo 

```python
def kde_statsmodels_m(x, x_grid, bandwidth=0.2, **kwargs):
    """Multivariate Kernel Density Estimation with Statsmodels"""
    kde = KDEMultivariate(x, bw = 'cv_ml',#bw=bandwidth * np.ones_like(x),
                          var_type='c', **kwargs)
    return kde.pdf(x_grid)




def kde_sklearn(x, x_grid, bandwidth=0.2, **kwargs):
    """Kernel Density Estimation with Scikit-learn"""
    kde_skl = KernelDensity(bandwidth=bandwidth, **kwargs)
    kde_skl.fit(x[:, np.newaxis])
    # score_samples() returns the log-likelihood of the samples
    log_pdf = kde_skl.score_samples(x_grid[:, np.newaxis])
    return np.exp(log_pdf)


kde_funcs = [kde_statsmodels_m, kde_sklearn]



def plot_Acep_rechazo(R2:'Variables distruidas U~U(0,1)',
                 R1:'Variables distribuidas como g(x)',
                 f:'función objetivo a generar',
                 t:'función que mayora a f'):
    f_x = f
    t_x = t
    condition = R2*t_x <= f_x
    [plt.plot(R1[i],R2[i]*t_x[i],'ob',markersize=1) if condition[i] else plt.plot(R1[i],R2[i]*t_x[i],'o',markersize=1) for i in range(len(R1))]         
    plt.show()
    return [R1[i] for i in range(len(R1)) if condition[i]]

    
    
    
def data_Acep_rechazo(R2:'Variables distruidas U~U(0,1)',
                 R1:'Variables distribuidas como g(x)',
                 f:'función objetivo a generar',
                 t:'función que mayora a f'):
    f_x = f
    t_x = t
    condition = R2*t_x <= f_x
    alea = [R1[i] for i in range(len(R1)) if condition[i]]                
    return alea

def plt_graficas(x_grid,pdf_stats,t,nombre):
    plt.plot(x_grid,pdf_stats,label='f(x)')
    plt.plot(x_grid,t,label='t(x)')
    plt.title(nombre)
```

#### Las aguilas

```python
# Aceptacion y Rechazo Aguilas 
N_1 = 1000
x_ca = np.asarray(lista_co[0][1:])
pdf_sk_cag = np.exp(kde_ag.score_samples(x_grid_ag[:, None]))

pdf_stats_cag = kde_statsmodels_m(x_ca, x_grid_ag, bandwidth=kde_ag.bandwidth)

f = pdf_stats_cag
# Máximo de la función f
max_f = np.max(f)
# Función t -> Función constante
t = max_f*np.ones(len(x_grid_ag)) # función constante

plt_graficas(x_grid_ag,pdf_stats_cag,t,'Co_Aguilas')

# Como estoy tomando t(x) constante solo es necesario generar valores aleatorios U~(0,1)
R2 = np.random.rand(N_1)
R1 = x_grid_ag
acep_cag=plot_Acep_rechazo(R2,R1,f,t)
```

```python
x_ta = np.asarray(lista_temp[0][1:])
pdf_sk_ta = np.exp(kde_tag.score_samples(x_grid_tag[:, None]))
pdf_stats_ta = kde_statsmodels_m(x_ta, x_grid_tag, bandwidth=kde_tag.bandwidth)
f_ta = pdf_stats_ta
# Máximo de la función f
max_f_ta = np.max(f_ta)
# Función t -> Función constante
t_ta = max_f_ta*np.ones(len(x_grid_tag)) # función constante

#Gráficas de las funciones
plt_graficas(x_grid_tag,pdf_stats_ta,t_ta,'Temp_Aguilas')
# Como estoy tomando t(x) constante solo es necesario generar valores aleatorios U~(0,1)
R2_ta = np.random.rand(N_1)
R1_ta = x_grid_tag
acep_tag=plot_Acep_rechazo(R2_ta,R1_ta,f_ta,t_ta)
```

### 1.8 Conclusiones

- Nuestra variable de estudio del CO sigue una distribución Poisson, mientras que la variable de estudio Temperatura sigue una distribución normal.

- Los anchos de banda en el caso del CO para las zonas de las águilas, centro y vallarta son iguales, pero para la zona de tlaquepaque es distinto. Esto por los rangos de CO presentes en esta zona en particulas ya que su histograma era el que menos parecido tenía con las demás zonas. 

- Los anchos de banda para el caso de la temperatura todos son distintos debido a los rangos que maneja cada zona. 

- Observamos como la máxima y mínima temperatura en cada zona es diferente a pesar de que la distancia entre ambas colonias no es extremadamente grande, por lo que diríamos que el calentamiento global afecta diferente de acuerdo a las zonas atribuyendo este fenómeno a la presencia de individuos que generan más contaminación.

- La correlación nos muestra que ambas variables presentan un porcentaje de correlación inversamente proporcional, es decir, mientras que una variable aumenta la otra disminuye y viceversa, sin embargo nuestra investigación previa nos indica que la temperatura se ve afectada en gran medida por este compuesto químico, pero gracias a esta prueba podemos decir que si bien es un factor que incluye en gran medida, la combinación con otros químicos los cuales estan presentes en el ambiente hacen que la temperatura incremente y no solo la cantidad de CO que pueda tener el aire de la zona en cuestión.

- Por medio de aceptación y rechazo pudimos generar variables aleatorias que siguieran la distribución de cada una de nuestras variables de estudio (CO y Temperatura) lo que nos permitió despues realizar el cálculo del promedio de estas estimaciones para así hacer una estimación final. La generación de variables aleatorias nos puede servir para así hacer estimaciones futuras en base a nuestros datos.
